```python
# Parameters
shape = (256, 256)
scale = 10.0
octaves = 6
persistence = 0.5
lacunarity = 2.0
seed = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
import noise

def generate_perlin_noise(shape=(256, 256), scale=10.0, octaves=1, seed=0):
    """
    Generate Perlin noise using the noise library.
    """
    noise_array = np.zeros(shape)
    for x in range(shape[0]):
        for y in range(shape[1]):
            noise_array[x][y] = noise.snoise2(x / scale, y / scale, repeatx=1024, repeaty=1024, base=seed)
    return noise_array

def display_noise(noise_array):
    """
    Display the generated noise using matplotlib.
    """
    plt.figure(figsize=(10, 10))
    plt.imshow(noise_array, cmap='gray')
    plt.colorbar()
    plt.title("Perlin Noise")
    plt.show()

# Generate and display Perlin noise
noise_data = generate_perlin_noise()
display_noise(noise_data)
```
